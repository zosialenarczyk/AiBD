```python
import matplotlib.pyplot as plt
import numpy as np

y = lambda x: x**2 + 5

x1 = np.arange(-1, 1, 0.01)
x2 = np.arange(-6, 6, 0.01)
x3 = np.arange(0, 5, 0.01)

def wykres(func, arg):
    plt.plot(func(arg))
    plt.title("Wykres funkcji")
    plt.legend()
    plt.xlabel("O x")
    plt.ylabel("O y")
    plt.show()

wykres(y, x1)
wykres(y, x2)
wykres(y, x3)
```
